````python
import csv#Library for reading csv format files
import csv

#Here we go!fileObj = open('budget_data.csv', newline='')
#reader is the cursor for the file
reader = csv.reader(fileObj)
#data will be written into a list with a 2D array so that the list
#Looks like this: [['StringDate','Profit/LossStringInteger'],['StringDate','Profit/LossStringInteger']]
data = list(reader)
print(data,'\n')

#Close resources that are no longer needed
fileObj.close()

newData = []
num_of_months = 0

#Iteration Loop for filling out newData list
for x in data[1:]:
    #newData [[Date, Profit/Loss],[Date, Profit/Loss],...,[Date, Profit/Loss][]
    #Switches Profit/Loss from Strings to Integers
    x[1]=int(x[1])
    #Attaches element to newData List
    newData.append(x)
    #Typecasting string integers into int types
    #newData[x][1] = int(data[x][1])
    #Keeps track of number of months and index for newData
    num_of_months += 1

#Checking integrity newData list
print(newData,'\n')


#Declarations for global Calculation variables
profitSum = 0
worstDeltaMonth = 0
worstDelta = 0
bestDeltaMonth = 0
bestDelta = 0
profitAverage = 0
deltaProfitSum = 0
prevProfit = 0

#Iterating through newData for all calculations
for i in newData:
    #Takes Profit/Losses of each element in newData and adds them into profitSum
    profitSum += i[1]
    #Checks whether start of newData list or not
    
    #Calculates the change in Profit/Losses between current month and the prior month
    monthDelta = i[1] - prevProfit
    #Adds the calculated change in monthly profit, and adds to total change in profit
    deltaProfitSum += monthDelta

    #Checks whether monthDelta is a greater decrease in profit than the saved worstDelta
    if(monthDelta<worstDelta):
        #Assigns month with worst change in profit to current iteration
        worstDeltaMonth = i
        #Assigns new saved worst change in profit to current month (iteration)
        worstDelta = monthDelta

        

    #Checks whether monthDelta is a greater increase in profit than saved bestDelta    
    elif(monthDelta>bestDelta):
        #Assigns month with best change in profit to current month (iteration)
        bestDeltaMonth = i
        bestDelta = monthDelta
        
#Prints to show new saved worstDelta
print('Total number of months: ', num_of_months,'\n')

print('Worst change in Profit: ', worstDelta,'\n')

print('Best Change in Profit: ', bestDelta,'\n')

print('Net profit/loss: ', profitSum,'\n')

deltaProfitAverage = deltaProfitSum / num_of_months
print('Average change in profit: ', deltaProfitAverage,'\n')

#Format and write to an output.txt text document
outFile = open('output.txt','w')

outFile.write('Financial Analysis\n')
outFile.write('---------------------\n')
outFile.write('Total Months: {}\n'.format(num_of_months))
outFile.write('Total: ${}\n'.format(profitSum))
outFile.write('Average Change: ${}\n'.format(deltaProfitAverage))
outFile.write('Greatest Increase in Profit: '+bestDeltaMonth[0]+' ($'+str(bestDelta)+')\n')
outFile.write('Greatest Decrease in Profit: '+worstDeltaMonth[0]+' ($'+str(worstDelta)+')\n')

print("Successfully written to output.txt\n")
#Close the text document, as it will no longer need to be edited in this iteration
outFile.close()
````
